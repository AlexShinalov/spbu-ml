```python
import numpy as np
import cv2
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import os
```

```python
image_size = (64, 64)
component = list(range(10, 201, 10))
var = []
accuracy = []
```

```python
def load_images_from_folder(folder, label):
    images = []
    labels = []
    for filename in os.listdir(folder):
        img_path = os.path.join(folder, filename)
        img = cv2.imread(img_path)
        if img is not None:
            img = cv2.resize(img, image_size)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            img = img.flatten()
            images.append(img)
            labels.append(label)
    return np.array(images), np.array(labels)
```

```python
train_cats, y_train_cats = load_images_from_folder(train_cats_dir, 0)
train_dogs, y_train_dogs = load_images_from_folder(train_dogs_dir, 1)
test_cats, y_test_cats = load_images_from_folder(test_cats_dir, 0)
test_dogs, y_test_dogs = load_images_from_folder(test_dogs_dir, 1)

X_train = np.vstack([train_cats, train_dogs])
y_train = np.hstack([y_train_cats, y_train_dogs])
X_test = np.vstack([test_cats, test_dogs])
y_test = np.hstack([y_test_cats, y_test_dogs])
```

```python
y_test.shape
```

```python
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

image_size = (64, 64)
component = list(range(1, 200, 10))
var = []
accuracy = []
"""
clf1 = SVC()
clf1.fit(X_train_scaled, y_train)
y_pred = clf1.predict(X_test_scaled)
acc = accuracy_score(y_test, y_pred)
print(f"Accur svc {acc}")

clf2 = LogisticRegression(max_iter=100)
clf2.fit(X_train_scaled, y_train)
y_pred = clf2.predict(X_test_scaled)
acc = accuracy_score(y_test, y_pred)
print(f"Accur LR {acc}")
"""
clf3 = DecisionTreeClassifier()
clf3.fit(X_train_scaled, y_train)
y_pred = clf3.predict(X_test_scaled)
acc = accuracy_score(y_test, y_pred)
print(f"Accur {acc}")

clf = clf3
for n in component:
    pca = PCA(n_components=n)
    X_train_pca = pca.fit_transform(X_train_scaled)
    X_test_pca = pca.transform(X_test_scaled)

    clf.fit(X_train_pca, y_train)
    y_pred = clf.predict(X_test_pca)

    explained_var = np.sum(pca.explained_variance_ratio_)
    acc = accuracy_score(y_test, y_pred)

    var.append(explained_var)
    accuracy.append(acc)

    print(f"Компонент: {n} | Точность: {acc:.3f} | дисперсия: {explained_var:.3f}")
```

```python
plt.figure(figsize=(10,5))
plt.plot(component, accuracy, marker='o')
plt.title("Зависимость accuracy от числа компонент PCA")
plt.xlabel("Число компонент")
plt.ylabel("Accuracy")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(component, var, marker='o', color='orange')
plt.axhline(0.9, color='red', linestyle='--', label='90% дисперсии')
plt.title("Дисперсия и комоненты")
plt.xlabel("число компонент")
plt.ylabel("дисперсия")
plt.legend()
plt.grid(True)
plt.show()
```

```python
pca_full = PCA(n_components=31)
pca_full.fit(X_train_scaled)

fig, axes = plt.subplots(1, 10, figsize=(15, 3))
for i in range(10):
    component_img = pca_full.components_[i].reshape(image_size)
    axes[i].imshow(component_img, cmap='gray')
    axes[i].axis('off')
    axes[i].set_title(f"PC {i+1}")
plt.suptitle("Первые 10 компонент PCA")
plt.tight_layout()
plt.show()
```
